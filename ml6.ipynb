```python
#importing necessary libraries
import numpy as np
import pandas as pd
```

```python
#importing the dataset
df = pd.read_csv("nba_2013.csv")
```

```python
from sklearn.preprocessing import Imputer
from sklearn.cross_validation import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import accuracy_score
```

```python
#columns to drop
#player- player name has no effect on no of points scored
#columns for which we have to create dummy are pos, bref_team_id,season
df.isnull().sum()
```

```python
#dropping the columns which have no impact
new_df = df.drop(columns=["player","season_end","season"])
```

```python
new_df.head()
```

```python
nominal_columns = ["pos","bref_team_id"]
```

```python
#creating dummy variables from the nominal columns
dummy_df = pd.get_dummies(new_df[nominal_columns])
```

```python
dummy_df
```

```python
#joining dummy columns and original data 
ml_df = pd.concat([new_df,dummy_df], axis=1)
```

```python
#dropping nominal columns that we have converted to dummy data
ml_df.drop(nominal_columns,axis=1,inplace=True)
```

```python
ml_df.head()
```

```python
#filling missing values with mean
ml_df.fillna(ml_df.mean(), inplace = True)
```

```python
#checking missing values
ml_df.isna().sum()
```

```python
ml_df.columns
```

```python
#dividing the data into features and labels 
X = ml_df[['age', 'g', 'gs', 'mp', 'fg', 'fga', 'fg.', 'x3p', 'x3pa', 'x3p.',
       'x2p', 'x2pa', 'x2p.', 'efg.', 'ft', 'fta', 'ft.', 'orb', 'drb', 'trb',
       'ast', 'stl', 'blk', 'tov', 'pf', 'pos_C', 'pos_F', 'pos_G',
       'pos_PF', 'pos_PG', 'pos_SF', 'pos_SG', 'bref_team_id_ATL',
       'bref_team_id_BOS', 'bref_team_id_BRK', 'bref_team_id_CHA',
       'bref_team_id_CHI', 'bref_team_id_CLE', 'bref_team_id_DAL',
       'bref_team_id_DEN', 'bref_team_id_DET', 'bref_team_id_GSW',
       'bref_team_id_HOU', 'bref_team_id_IND', 'bref_team_id_LAC',
       'bref_team_id_LAL', 'bref_team_id_MEM', 'bref_team_id_MIA',
       'bref_team_id_MIL', 'bref_team_id_MIN', 'bref_team_id_NOP',
       'bref_team_id_NYK', 'bref_team_id_OKC', 'bref_team_id_ORL',
       'bref_team_id_PHI', 'bref_team_id_PHO', 'bref_team_id_POR',
       'bref_team_id_SAC', 'bref_team_id_SAS', 'bref_team_id_TOR',
       'bref_team_id_TOT', 'bref_team_id_UTA', 'bref_team_id_WAS']]
Y= ml_df["pts"]
```

```python
#didvind the data in to train and test
X_train, X_test, y_train, y_test = train_test_split(
 X, Y, test_size = 0.3, random_state = 100)
```

```python
#fitting the model finding the accuracy
for K in range(25):
    K_value = K+1
    neigh = KNeighborsRegressor(n_neighbors = K_value, weights='uniform', algorithm='auto')
    neigh.fit(X_train, y_train) 
    y_pred = neigh.predict(X_test)
    #print("neigh.score(X_test,y_test)")
    print("Accuracy is ",neigh.score(X_test,y_test) ," for K-Value:",K_value)
```
